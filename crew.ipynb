```python
from crewai import Agent, Task, Crew
```

```python
import os

os.environ['OPENAI_API_KEY'] = "MY_API_KEY"
os.environ['OPENAI_MODEL_NAME'] = 'gpt-3.5-turbo'
```

Grupo de agentes para escrever artigos

1. Panner - Planejador do artigo
2. Writer - Escritor do artigo
3. Editor - Editor do artigo

```python
planner = Agent(
    role="Content planner",
    goal="Plan engaging and factually accurate content about the {topic}",
    backstory="""You're working on planning a blog post article about 
    the topic: {topic},
    You collect information that hepls the audience learn something 
    and make informed decisions.
    Tou work is the basis for the Content Writer to write the article on this topic
    """,
    allow_delegation=False,
    verbose=True
)
```

```python
writer = Agent(
    role="Content writer",
    goal="Write insightfull and factually accurate opinion piece about the topic: {topic}",
    backstory="""You're working on writing an opinion about the topic: {topic}
    You base your opinion on the work of the Content Planner, who provides an
    outline an relevant context about the topic. You folow the main objectives and 
    direction of the outline, as provided by the Content Planner.
    You also provide objective and impartial insights and back them with information provided
    by the Content Planner
    """,
    allow_delegation=False,
    verbose=True
)
```

```python
editor = Agent(
    role="Editor",
    goal="Edit a given blog post article to align with the writing style of the organization",
    backstory="""You are a editor that receives a blog post from the Content Writer
    Your goal is to review the blog post to ensure that it follow journalistic best practices.
    """,
    allow_delegation=False,
    verbose=True
)
```

```python
plan = Task(
    description=(
        "1. Prioritaze the last trends, key players, and noteworthy news on {topic}"
        "2. Identify the target audience, considering their interests and pain points"
        "3. Develop a detaild content outline including an introduction, key points and call to action."
        "4. Include SEO keyword and relevant data or sources."
    ),
    expected_output="A comprehensive content plan document with an outline, audience analysis, SEO keywords and resources.",
    agent=planner
)
```

```python
write = Task(
    description=(
        "1. Use the content plan to craft a compelling blog post on topic: {topic}."
        "2. Incorporate the SEO keywods naturally."
        "3. Sections/Subtitles are properly named in engaging manner."
        "4. Ensure the post is sctructured with an engaging introduction, insightfull body and summarization conclusion."
    ),
    expected_output="A well-writen blog post in Markdown format, ready fo publication, each section should have 2 or 3 paragraphs",
    agent=writer
)
```

```python
edit = Task(
    description=("Proofread the given blog post for grammatical errors and alignment with the brand's voice"),
    expected_output="A well-writen blog post in Markdown format, ready fo publication, each section should have 2 or 3 paragraphs",
    agent=editor
)
```

```python
crew = Crew(
    agents=[planner, writer, editor],
    tasks=[plan, write, edit],
    verbose=2
)
```

```python
result = crew.kickoff(inputs={"topic": "Artificial Inteligence"})
```
